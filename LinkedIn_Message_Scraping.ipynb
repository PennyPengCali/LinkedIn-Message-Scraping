```python
#pip install selenium

#load the library
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
```

```python
#define driver
driver =webdriver.Chrome('/Users/pengshuang/Downloads/chromedriver')
driver.get("https://www.linkedin.com")

#waits for opening the web page
time.sleep(2)

#get the element of input and output
username= driver.find_element(by=By.XPATH, value="//input[@name='session_key']")
password= driver.find_element(by=By.XPATH, value="//input[@name='session_password']")

#input username and keyword(I have deleted my own username and password)
username.send_keys('********')
password.send_keys('********')

time.sleep(2)

#get the element of submit and click submit
submit= driver.find_element(by=By.XPATH, value="//button[@type='submit']").click()
time.sleep(2)
```

```python
#30 buttons to click 
chat_button_list = driver.find_elements(by=By.CLASS_NAME, value="msg-conversation-listitem__link")[:30]
```

```python
#get element of message list
js_top = 'document.getElementsByClassName("msg-s-message-list")[0].scrollTop=-10000'

#get all the messages
message = []
for chat_button in chat_button_list:
    chat_button.click()
    #wait for the chat box to open
    time.sleep(2)
    
    #scroll to the top of the chat box
    driver.execute_script(js_top)
    
    #get alll the messages for 30 conversation using a for loop
    chat_box = driver.find_element(by=By.CLASS_NAME, value="msg-s-message-list") 
    chat_item = chat_box.find_element(by=By.CLASS_NAME, value="msg-s-message-list-content")
    chat=chat_item.find_elements(by=By.CLASS_NAME, value="msg-s-message-list__event")
    for c in chat:
        message.append(c.text)
        time.sleep(2)
    print(message)
    mes_box=driver.find_element_by_class_name("msg-form__contenteditable")
    mes_box.send_keys(Keys.ESCAPE)
    time.sleep(2)
```
